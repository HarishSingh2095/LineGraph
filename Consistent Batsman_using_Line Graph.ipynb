## A simple Line Graph to understand which player is consistent in 4 ODIs played previously:

### Step 1: Import the necessary modules

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

### Step 2: Create a necessary DataFrame of the batsmen scores:

```python
data = {'batsman1':[97, 20, 0, 60],
        'batsman2':[10, 0, 10, 140],
        'batsman3':[10, 65, 75, 30],
       'Match':["Odi1", "Odi2", "Odi3", "Odi4"]}
df = pd.DataFrame(data)
```

```python
df
```

### Step 3: We will calculate Mean and Std Deviation of the data to find out consistency statistically

```python
print(df.mean())
```

```python
print(df.std())
```

### Step 4: Finally plot a Line Graph to understand the performance of Batsmen graphically

```python
df.plot(kind='line', figsize=(10,6))
plt.xlabel('ODIs')
plt.ylabel('Scores')
plt.show()
```

### Observation: 
Here we can see that Batsman 3 has been more consistent in all the 4 ODIs played, as Batsman3 has never scored 0, though he has not been one of the highest scorer he has been contributing consistently for his team in all the matches played.

Here the mean of Batsman3 = 45.00 is highest and also Standard Deviation = 30.276504 which is least deviating from other batsmen performances.
